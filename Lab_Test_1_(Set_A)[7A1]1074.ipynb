```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

scores = np.array([93, 83, 69, 62, 75, 93, 73, 77, 65, 65,
                   78, 90, 94, 78, 57, 76, 56, 78, 84, 92])

def basic_stats(data):
    mean_val = np.mean(data)
    pop_var = np.var(data)           # population variance (ddof=0)
    sample_var = np.var(data, ddof=1)  # sample variance (ddof=1)
    pop_std = np.std(data)           # population std deviation
    sample_std = np.std(data, ddof=1)  # sample std deviation

    print("Basic Statistics:")
    print(f"Mean: {mean_val:.2f}")
    print(f"Population Variance: {pop_var:.2f}")
    print(f"Sample Variance: {sample_var:.2f}")
    print(f"Population Std Dev: {pop_std:.2f}")
    print(f"Sample Std Dev: {sample_std:.2f}\n")

basic_stats(scores)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# --- 3. Monte Carlo Simulation ---
np.random.seed(42)  # For reproducibility

samples = np.random.normal(loc=mean_score, scale=pop_std, size=50000)
prob_ge_75 = np.mean(samples >= 75)
print(f"Monte Carlo Simulation: Probability that score >= 75 = {prob_ge_75:.4f}\n")

# Optional visualization of Monte Carlo samples highlighting scores >= 75
plt.figure(figsize=(8, 4))
plt.hist(samples, bins=50, color='purple', alpha=0.7, edgecolor='black')
plt.axvline(75, color='red', linestyle='--', label='Score = 75')
plt.fill_betweenx([0, plt.gca().get_ylim()[1]], 75, samples.max(), color='red', alpha=0.2)
plt.title("Distribution of Monte Carlo Simulated Scores")
plt.xlabel("Score")
plt.ylabel("Frequency")
plt.legend()
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()



```

```python
#  4. Hypothesis Testing
student_id_last_three_digits = 174
hypothesized_mean = (student_id_last_three_digits % 50) + 50

t_stat, p_value = stats.ttest_1samp(scores, hypothesized_mean)

print("Hypothesis Testing:")
print(f"Hypothesized Mean: {hypothesized_mean}")
print(f"t-statistic: {t_stat:.4f}")
print(f"p-value (two-tailed): {p_value:.4f}")

if p_value < 0.05:
    print("Conclusion: Reject the null hypothesis (significant difference).")
else:
    print("Conclusion: Fail to reject the null hypothesis (no significant difference).")

# Visualization: sample mean vs hypothesized mean
plt.figure(figsize=(8, 5))
plt.hist(scores, bins=8, color='yellow', alpha=0.7, edgecolor='red', label="Exam Scores")
plt.axvline(mean_score, color='blue', linestyle='--', linewidth=2, label=f'Sample Mean = {mean_score:.2f}')
plt.axvline(hypothesized_mean, color='red', linestyle='--', linewidth=2, label=f'Hypothesized Mean = {hypothesized_mean}')
plt.title("Hypothesis Testing: Sample Mean vs Hypothesized Mean")
plt.xlabel("Scores")
plt.ylabel("Frequency")
plt.legend()
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()
```
